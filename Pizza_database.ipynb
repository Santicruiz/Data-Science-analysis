```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar los datos
df = pd.read_csv('pizza_sales.csv')
```

```python
# Pregunta exploratoria 1: ¿Existe una correlación entre el tamaño de la pizza y su precio?
# Se crea un gráfico de dispersión para visualizar la relación entre el tamaño de la pizza (en el eje x) y su precio (en el eje y).
# La dispersión de puntos ayuda a identificar patrones o tendencias en la relación entre estas dos variables.
# Si los puntos tienden a agruparse en una dirección específica, podría indicar una correlación entre el tamaño de la pizza y su precio.
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='pizza_size', y='unit_price')
plt.title('Correlación entre el tamaño de la pizza y su precio')
plt.show()
```

```python
# Pregunta exploratoria 2: ¿Las pizzas con más ingredientes son más caras?
# Se crea una nueva columna 'num_ingredients' que cuenta el número de ingredientes en cada pizza.
# Luego, se utiliza un gráfico de barras para visualizar el precio promedio de las pizzas en función del número de ingredientes.
# Este gráfico ayuda a entender si hay una relación entre la complejidad (número de ingredientes) de la pizza y su precio.
df['num_ingredients'] = df['pizza_ingredients'].str.split(',').str.len()
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='num_ingredients', y='unit_price')
plt.title('Precio promedio de las pizzas por número de ingredientes')
plt.show()
```

```python
# Pregunta exploratoria 3: ¿Existe una relación entre la hora del día y el tipo de pizza ordenada?
# Se convierten las horas de la columna 'order_time' al formato de 24 horas utilizando pd.to_datetime() y dt.hour.
# Luego, se crea un gráfico de barras apiladas para mostrar la cantidad de cada tipo de pizza ordenada en diferentes horas del día.
# Se ajusta la posición y la rotación de las etiquetas en el eje x para mejorar la visualización.
plt.figure(figsize=(12, 8))  # Ajusta el tamaño de la figura
sns.countplot(data=df, x='order_time', hue='pizza_name')
plt.title('Cantidad de cada tipo de pizza ordenada por hora del día')
plt.legend(title='Tipo de Pizza', loc='upper right', bbox_to_anchor=(1.4, 1))
plt.xlabel('Hora del Día')
plt.ylabel('Cantidad de Pizzas Ordenadas')
plt.xticks(rotation=45, ha='right')  # Ajusta la rotación de las etiquetas en el eje x
plt.show()


```

```python
# Pregunta exploratoria 4: ¿Existe una correlación entre la categoría de la pizza y su precio?
# Se utiliza un gráfico de violín para visualizar la distribución de los precios de las pizzas en cada categoría.
# Este gráfico muestra la densidad de la distribución 
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='pizza_category', y='unit_price')
plt.title('Distribución de precios por categoría de pizza')
plt.show()
```

```python
# Pregunta exploratoria 5: ¿Las pizzas con ciertos ingredientes se venden más que otras?
# Se realiza un conteo del número de pizzas que contienen cada ingrediente, desglosando la lista de ingredientes.
# Luego, se crea un gráfico de barras para visualizar la frecuencia de cada ingrediente en las pizzas vendidas.
# Este gráfico proporciona información sobre la popularidad de los ingredientes y ayuda a identificar tendencias en las preferencias.
ingredient_counts = df['pizza_ingredients'].str.split(',').explode().value_counts()
ingredient_counts.plot(kind='bar', figsize=(12, 6))
plt.title('Número de pizzas vendidas por ingrediente')
plt.show()
```

```python
# Pregunta exploratoria 7: ¿Existe una correlación entre el nombre de la pizza y su precio?
# Se utiliza un gráfico de punto (strip plot) para visualizar la distribución de los precios de las pizzas en función de su nombre.
# Cada punto representa el precio de una pizza, y los puntos se distribuyen a lo largo del eje x según el nombre de la pizza.
plt.figure(figsize=(14, 6))
sns.stripplot(data=df, x='pizza_name', y='unit_price', jitter=True, alpha=0.7)
plt.title('Distribución de precios por nombre de pizza')
plt.xlabel('Nombre de Pizza')
plt.ylabel('Precio')
plt.xticks(rotation=45, ha='right')  # Rotamos los nombres de las pizzas para mejorar la legibilidad
plt.show()
```

```python
#¿Cuántos pedidos se esperan para los próximos siete días basándose en las tendencias actuales?

import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
data = pd.read_csv('pizza_sales.csv')

# Convertir la columna 'order_date' a tipo datetime
data['order_date'] = pd.to_datetime(data['order_date'], errors='coerce')

# Calcular el número de pedidos diarios
daily_orders = data.groupby('order_date').size().reset_index(name='num_orders')

# Crear una nueva columna 'day_number' que representa el número de días transcurridos
daily_orders['day_number'] = (daily_orders['order_date'] - daily_orders['order_date'].min()).dt.days

# Graficar el número de pedidos diarios
plt.figure(figsize=(10, 6))
plt.scatter(daily_orders['day_number'], daily_orders['num_orders'])
plt.title('Número de Pedidos Diarios')
plt.xlabel('Días Transcurridos')
plt.ylabel('Número de Pedidos')
plt.show()

# Entrenar un modelo de regresión lineal simple
model = LinearRegression()
model.fit(daily_orders[['day_number']], daily_orders['num_orders'])

# Graficar la regresión lineal
plt.figure(figsize=(10, 6))
plt.scatter(daily_orders['day_number'], daily_orders['num_orders'], label='Datos Observados')
plt.plot(daily_orders['day_number'], model.predict(daily_orders[['day_number']]), color='red', label='Regresión Lineal')
plt.title('Regresión Lineal sobre el Número de Pedidos Diarios')
plt.xlabel('Días Transcurridos')
plt.ylabel('Número de Pedidos')
plt.legend()
plt.show()

```

```python
# ¿Hay días específicos de la semana o horas del día en los que se realizan pedidos más grandes?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from dateutil import parser

# Función para intentar analizar la fecha y hora
def parse_datetime(date_str):
    try:
        return parser.parse(date_str)
    except:
        return None

# Aplicar la función a las columnas 'order_date' y 'order_time'
data['order_datetime'] = data.apply(lambda row: parse_datetime(row['order_date'] + ' ' + row['order_time']), axis=1)

# Verificar si hay valores nulos en la nueva columna
if data['order_datetime'].isnull().any():
    raise ValueError("La columna 'order_datetime' no se creó correctamente debido a problemas con los formatos de fecha y hora.")

# Continuar con el resto del código
orders_by_day = data.groupby(data['order_datetime'].dt.day_name()).size()
plt.figure(figsize=(10, 6))
sns.barplot(x=orders_by_day.index, y=orders_by_day.values, hue=orders_by_day.index, palette="viridis", legend=False)
plt.title('Número de Órdenes por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Número de Órdenes')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil import parser

# Cargar los datos desde el archivo CSV
data = pd.read_csv('pizza_sales.csv')

# Función para intentar analizar la fecha y hora
def parse_datetime(date_str):
    try:
        return parser.parse(date_str)
    except:
        return None

# Aplicar la función a las columnas 'order_date' y 'order_time'
data['order_datetime'] = data.apply(lambda row: parse_datetime(row['order_date'] + ' ' + row['order_time']), axis=1)

# Filtrar filas con fechas y horas válidas
data = data.dropna(subset=['order_datetime'])

# Establecer 'order_datetime' como índice
data.set_index('order_datetime', inplace=True)

# Calcular los ingresos mensuales
monthly_revenue = data.resample('M')['total_price'].sum()

# Graficar
plt.figure(figsize=(10, 6))
monthly_revenue.plot(marker='o')
plt.title('Ingresos Mensuales')
plt.xlabel('Fecha')
plt.ylabel('Ingresos Totales')
plt.show()
```

```python
#¿Cuántos pedidos se esperan para los próximos siete días basándose en las tendencias actuales?
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Cargar los datos desde el archivo CSV
data = pd.read_csv('pizza_sales.csv')

# Función para intentar analizar la fecha
def parse_date(date_str):
    try:
        return pd.to_datetime(date_str, format="%m/%d/%Y")
    except:
        return pd.NaT

# Aplicar la función a la columna 'order_date'
data['order_date'] = data['order_date'].apply(parse_date)

# Filtrar filas con fechas válidas
data = data.dropna(subset=['order_date'])

# Calcular el número de pedidos diarios
daily_orders = data.groupby('order_date').size()

# Crear un DataFrame con el índice como días y una columna para el número de pedidos
daily_orders_df = pd.DataFrame({'date': daily_orders.index, 'orders': daily_orders.values})

# Ajustar un modelo de regresión lineal
model = LinearRegression()
model.fit(np.array(daily_orders_df.index).reshape(-1, 1), daily_orders_df['orders'])
trend_line = model.predict(np.array(daily_orders_df.index).reshape(-1, 1))

# Crear un gráfico de líneas con la tendencia
plt.figure(figsize=(10, 6))
plt.plot(daily_orders_df['date'], daily_orders_df['orders'], marker='o', label='Número de Pedidos Diarios')
plt.plot(daily_orders_df['date'], trend_line, label='Tendencia', linestyle='--', color='red')
plt.title('Número de Pedidos Diarios y Tendencia')
plt.xlabel('Fecha')
plt.ylabel('Número de Pedidos')
plt.legend()
plt.show()
```

```python
#¿Se espera un cambio en la preferencia de tamaño de pizza en los próximos meses?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dateutil import parser

# Cargar los datos desde el archivo CSV
data = pd.read_csv('pizza_sales.csv')

# Convertir la columna 'order_datetime' a tipo datetime
data['order_datetime'] = data.apply(lambda row: parser.parse(f"{row['order_date']} {row['order_time']}"), axis=1)

# Crear una nueva columna con el mes y año
data['month_year'] = data['order_datetime'].dt.to_period('M')

# Contar el número de pedidos para cada tamaño en cada mes
monthly_size_counts = data.groupby(['month_year', 'pizza_size']).size().unstack(fill_value=0)

# Graficar el tamaño de pizza pedidos a lo largo de los meses
plt.figure(figsize=(12, 6))
monthly_size_counts.plot(marker='o')
plt.title('Tamaños de Pizza Pedidos a lo largo de los Meses')
plt.xlabel('Mes y Año')
plt.ylabel('Número de Pedidos')
plt.legend(title='Tamaño de Pizza', loc='upper left')
plt.show()


```

```python
#¿Cuántos pedidos se esperaría para una nueva pizza en función de los ingredientes y la categoría de pizza?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos desde el archivo CSV
data = pd.read_csv('pizza_sales.csv')

# Crear una nueva pizza con ingredientes y categoría específicos
nueva_pizza = pd.DataFrame({
    'pizza_id': [999],
    'pizza_name': ['Nueva Pizza'],
    'quantity': [1],
    'order_date': ['1/1/2025'],
    'order_time': ['12:00:00'],
    'unit_price': [15.99],
    'total_price': [15.99],
    'pizza_size': ['M'],
    'pizza_category': ['Nueva Categoria'],
    'pizza_ingredients': ['Ingrediente1, Ingrediente2, Ingrediente3'],
    'pizza_name_id': [999],
})

# Agregar la nueva pizza a los datos existentes
data_with_new_pizza = pd.concat([data, nueva_pizza])

# Contar el número de pedidos para cada pizza en función de los ingredientes y la categoría
pizza_popularity = data_with_new_pizza.groupby(['pizza_category', 'pizza_ingredients']).size().reset_index(name='num_orders')

# Ordenar por número de pedidos en orden descendente
pizza_popularity = pizza_popularity.sort_values(by='num_orders', ascending=False)

# Graficar la popularidad de las pizzas existentes y la nueva pizza
plt.figure(figsize=(12, 6))
sns.barplot(x='pizza_category', y='num_orders', data=pizza_popularity, hue='pizza_ingredients', palette='viridis', dodge=False)
plt.title('Popularidad de Pizzas por Categoría e Ingredientes')
plt.xlabel('Categoría de Pizza')
plt.ylabel('Número de Pedidos')

# Mover la leyenda a la derecha y fuera del área del gráfico
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Cargar los datos desde el archivo CSV
data = pd.read_csv('pizza_sales.csv')

# Definir una función para convertir a datetime con varios formatos
def convert_to_datetime(date_str):
    formats = ["%d-%m-%Y %H:%M:%S", "%m/%d/%Y %H:%M:%S"]
    for fmt in formats:
        try:
            return pd.to_datetime(date_str, format=fmt)
        except ValueError:
            pass
    return pd.NaT

# Intentar convertir las columnas 'order_date' y 'order_time' a tipo datetime
data['order_datetime'] = data.apply(lambda row: convert_to_datetime(f"{row['order_date']} {row['order_time']}"), axis=1)

# Verificar si hay valores nulos en la nueva columna
if data['order_datetime'].isnull().any():
    raise ValueError("La columna 'order_datetime' no se creó correctamente debido a problemas con los formatos de fecha y hora.")

# Agregar una columna para la estación del año
data['season'] = data['order_datetime'].dt.month.map({1: 'Invierno', 2: 'Invierno', 3: 'Primavera', 4: 'Primavera',
                                                      5: 'Primavera', 6: 'Verano', 7: 'Verano', 8: 'Verano',
                                                      9: 'Otoño', 10: 'Otoño', 11: 'Otoño', 12: 'Invierno'})

# Contar la cantidad de pedidos en cada estación
orders_by_season = data['season'].value_counts()

# Graficar el diagrama de torta
plt.figure(figsize=(8, 8))
plt.pie(orders_by_season, labels=orders_by_season.index, autopct='%1.1f%%', startangle=90, colors=['springgreen', 'gold', 'coral', 'lightskyblue'])
plt.title('Distribución de Pedidos por Estación del Año')
plt.show()

```
